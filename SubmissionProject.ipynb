```python
from tkinter import * 
from tkinter.filedialog import asksaveasfile
import sqlite3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from matplotlib.figure import Figure
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
from sklearn.neighbors import KNeighborsClassifier

sql = None

def save_file():
    Target_path = Target_Database.get()    # gets the target path to be saved at 
    f = asksaveasfile(initialfile = Target_path,
    defaultextension=".txt", filetypes=[("All Files", "*"),("Text Documents", "*.txt")])

def SaveToTarget(df):
    global sql
    Target_Path = Target_Database.get()   # gets the target path 
    Targetsql = sqlite3.connect(Target_Path)  #connects to that database
    df.to_sql("NewTable", Targetsql)        # transforms the df into a database
    Targetsql.commit()                     #commits change
    
    
    
def LocalDatabase():   # need to add save to database feature
    
    global sql
    local_path = Local_Database.get()  #gets the path from the ui input 
    sql = sqlite3.connect(local_path) #sets the path to the connection
    
    #runs the necessary functions
    InputData()
    ShowData()
    
            
    #closes connection
def ShowData():
    global sql
    c= sql.cursor()
    Database = "SELECT * FROM Marketing_Campaign_Data"
    c.execute(Database)
    Database = c.fetchall()
    column_names = [col_names[0] for col_names in c.description]
    df = pd.DataFrame(Database, columns=column_names)
    
    

    
    
def InputData():
    
    global sql
    c = sql.cursor()
    DataRows = "SELECT * FROM Marketing_Campaign_Data LIMIT 0" # sets the selction to 0 rows so it is has no entries
    c.execute(DataRows)
    column_names = [names[0] for names in c.description]   # https://docs.python.org/3/library/sqlite3.html#cursor-objects , returns name of columns from last query
    for i in column_names:    # inserts each column name into the listbox 
        Lb1.insert(END, i)
        Lb1.grid()
        Lb2.insert(END, i)
        Lb2.grid()
        
        
#gets the user selected column names and can explore data with i
def TransformLR():
    
    global sql
    
    #gets the selected from list
    x = Lb1.get(ANCHOR)    
    y = Lb2.get(ANCHOR)
    
    Transform = f"SELECT `{x}`, `{y}` FROM Marketing_Campaign_Data"  #uses the selected from list as query arguements
    c = sql.cursor()
    c.execute(Transform)
    Columns = c.fetchall()
    
    #creates dataframe from the listbox selected and the rows associated with them
    df = pd.DataFrame(Columns, columns=[x, y])
    
    #drops any null values.
    df = df.dropna()
    
    LR(df,x, y)
def TransformClass():
    
    global sql
    x = Lb1.get(ANCHOR)
    y = Lb2.get(ANCHOR)
    
    Transform = f"SELECT `{x}`, `{y}` FROM Marketing_Campaign_Data"
    c = sql.cursor()
    c.execute(Transform)
    Columns = c.fetchall()
    
    #creates dataframe from the listbox selected and the rows associated with them
    df = pd.DataFrame(Columns, columns=[x, y])
    
    #drops any null values.
    df = df.dropna()
    
    #check if data is numerical.

    print(df)
    
    classifcation(df, x, y)
    
def TransformClustering():
    
    global sql
    x = Lb1.get(ANCHOR)
    y = Lb2.get(ANCHOR)
    
    Transform = f"SELECT `{x}`, `{y}` FROM Marketing_Campaign_Data"
    c = sql.cursor()
    c.execute(Transform)
    Columns = c.fetchall()
    
    #creates dataframe from the listbox selected and the rows associated with them
    df = pd.DataFrame(Columns, columns=[x, y])
    #drops any null values.
    df = df.dropna()
    

    print(df)
    clustering(df, x, y)

    
    
# use the transformed selected now in the MLAs

def LR(df, x, y): 
     
    x2 = df[x]
    y2 = df[y]
    #training the model
    x_train, x_test, y_train, y_test = train_test_split(x2, y2, test_size=0.3, random_state=0)
    LR = LinearRegression()
    LR.fit(x_train.values.reshape(-1, 1), y_train.values)
    
    prediction = LR.predict(x_test.values.reshape(-1, 1))

    #plotting the model
    plt.plot(x_test, prediction, label="Linear Regression", color="red")
    plt.scatter(x_test, y_test, label="Test Data", color="blue")
    plt.title("LR prediction Model between 2 chosen columns")
    plt.xlabel(x)
    plt.ylabel(y)
    plt.show()
    
    plot = Toplevel(root)
    plot.title("LR prediction Model")
    plot.geometry("400x400")
    
    #plotting in canvas
    fig = Figure(figsize=(5, 5), dpi=75)
    a = fig.add_subplot(111)
    a.plot(x_test, prediction, label="Linear Regression", color="red")
    a.scatter(x_test, y_test, label="Test Data", color="blue")
    a.set_title("LR prediction Model between 2 chosen columns")  
    a.set_xlabel(x)  
    a.set_ylabel(y)  

    canvas = FigureCanvasTkAgg(fig, master=plot)
    canvas.get_tk_widget().grid()  
    canvas.draw()
    
    SaveToTarget(df)

def classifcation(df, x, y):
    
    x2 = df[x]
    x2 = x2.values.reshape(-1,1)
    y2 = df[y]
    
    # using Knearest algorithm
    #trains model
    x_train2, x_test2, y_train2, y_test2 = train_test_split(x2,y2, random_state = 0)
    scaler = MinMaxScaler()
    scaler.fit(x_train2)
    x_train2 = scaler.transform(x_train2)
    x_test2 = scaler.transform(x_test2)
    
    knn = KNeighborsClassifier(n_neighbors=3)   
    knn.fit(x_train2, y_train2) 
    y_pred = knn.predict(x_test2)
    
    #plots model
    plt.scatter(x_test2, y_test2, c=y_pred)
    plt.title("Classification between chosen columns")
    plt.xlabel(x)
    plt.ylabel(y)
    plt.show()
    
    plot = Toplevel(root)
    plot.title("LR prediction Model")
    plot.geometry("400x400")
    
    #plotting in canvas
    fig = Figure(figsize=(5, 5), dpi=75)
    a = fig.add_subplot(111)
    a.scatter(x_test2, y_test2, label="Test Data", c=y_pred)
    a.set_title("Classification Model between 2 chosen columns")  
    a.set_xlabel(x)  
    a.set_ylabel(y)  

    canvas = FigureCanvasTkAgg(fig, master=plot)
    canvas.get_tk_widget().grid()  
    canvas.draw()
    
    SaveToTarget(df)
    
def clustering(df, x, y):
    
    x2 = df[x]
    x2 = x2.values.reshape(-1,1)
    y2 = df[y]
    
    #trains the model
    km = KMeans(n_clusters=4)
    y_pred_clust = km.fit_predict(df[[x, y]])
    df["Cluster"] = y_pred_clust
    
    #plots the model
    plt.scatter(x=df[x], y=df[y], c=df["Cluster"])
    plt.title("Kmeans Clustering")
    plt.xlabel(x)
    plt.ylabel(y)
    plt.show()
    
    plot = Toplevel(root)
    plot.title("Clustering Model")
    plot.geometry("400x400")
    
    #plotting in canvas
    fig = Figure(figsize=(5, 5), dpi=75)
    a = fig.add_subplot(111)
    a.scatter(x=df[x], y=df[y], label="Test Data", c=df["Cluster"])
    a.set_title("Clustering Model between 2 chosen columns")  
    a.set_xlabel(x)  
    a.set_ylabel(y)  

    canvas = FigureCanvasTkAgg(fig, master=plot)
    canvas.get_tk_widget().grid()  
    canvas.draw()
    
    SaveToTarget(df)
    
root = Tk()
root.title("Marketing Data")
root.geometry("400x400")
root.configure(bg='PeachPuff4')


#local database 
#create text boxes
Local_Database = Entry(root, width=30, bg="PeachPuff3")
Local_Database.grid(row=0, column=1)
#create text box label
Local_Database_label = Label(root, text="Local Database Path", bg="PeachPuff3")
Local_Database_label.grid(row=0, column=0)

#target database

#create text boxes
Target_Database = Entry(root, width=30, bg="PeachPuff3")  # copy entry into the untitled spot
Target_Database.grid(row=1, column=1)
#create text box label
Target_Database_label = Label(root, text="Target Database Path", bg="PeachPuff3")
Target_Database_label.grid(row=1, column=0)

#button to submit local path
submit_button = Button(root, text="Submit",bg="PeachPuff3", command=LocalDatabase)
submit_button.grid(row=1, column=0, columnspan=1, pady=0, padx=0, ipadx=0) 
submit_button.place(x=310 , y=0)

#Selection for columns
Lb1 = Listbox(root, bg="PeachPuff3")  # https://www.youtube.com/watch?v=wEv3BworNK8 reference for listboxes
Lb1.place(x=0, y=100)
Lb2 = Listbox(root, bg="PeachPuff3")
Lb2.place(x=0, y=250)
#linear regression button
Reg_button = Button(root, text="Linear Regression", bg="PeachPuff3", command=TransformLR) 
Reg_button.grid(row=10, column=0, columnspan=2, pady=5, padx=10, ipadx=5)
Reg_button.place(x=132, y= 367)
#classification button
cla_button = Button(root, text="Classification", bg="PeachPuff3",command=TransformClass)
cla_button.grid(row=10, column=0, columnspan=2, pady=5, padx=10, ipadx=5)
cla_button.place(x=238, y= 367)
#clustering button
clus_button = Button(root, text="Clustering", bg="PeachPuff3",command=TransformClustering) 
clus_button.grid(row=10, column=0, columnspan=2, pady=5, padx=10, ipadx=5)
clus_button.place(x=323, y= 367)
#DataFrame display 
#df_display = Text(root, width=33, height=15, bg="PeachPuff3")
#df_display.grid(row=10, column=0, columnspan=2, pady=5, padx=10, ipadx=5)
#df_display.place(x=128, y=100)
#save button
save_button = Button(root, text="Save",width=6, command= lambda:save_file(), bg="PeachPuff3")
save_button.grid()
save_button.place(x=335, y=320)
root.mainloop()



#conn = sqlite3.connect(r"C:\Users\smith\Sql\DataBase1.db")



```
